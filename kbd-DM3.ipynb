```python
from bs4 import BeautifulSoup
import requests
import operator
```

```python
# url of top contributors
url = 'https://gist.github.com/paulmillr/2657075'
# limit number of contributors of interest
ncontributors = 256
# url of git API
git_url='https://api.github.com'
# limit number of REPOS considered (as a training parameter, to avoid lengthy computations)
limit_repo = 1
# my token number
token = '614a0903c78bd99c86cf448bd6e0ff7aaf39c825'
# to start the API
headers = {'Authorization': 'token ' + token}

# method to compute average star rating of REPO
def average_stars(repos):
    stars = 0
    count = 0
    star_dict = {}
    star_list = []
    for repo in repos:
        stars+=repo['stargazers_count']
        count+=1
        star_dict[repo['name']]=repo['stargazers_count'] # if needed
        star_list.append(repo['stargazers_count']) # if needed
    return stars/count    

# first obtain the names of the top 256 contributors
content = requests.get(url)
html = content.text
soup = BeautifulSoup(html, "html.parser")
# the scraping strategy is simple: the contributors are found in 'th' tags, with '#n' attributes
top_contributors=[]
for i in range (1,ncontributors+1,1):
    hash = '#'+str(i)
    contributor = str(soup.find('th', text=hash).find_next_sibling().find('a').contents[0])
    top_contributors.append(contributor)
print(top_contributors)


# Now obtain the star rating

star_dict = {}
for contributor in top_contributors:
    # get repos for contributor
    repos = requests.get(git_url+'/users/'+contributor+'/repos'+'?per_page='+str(limit_repo), headers=headers).json()
    if(len(repos)==0):
        print(("User {} has no repo!").format(contributor))
        continue
    print("{} repos found for {}".format(len(repos),contributor))
    mean_contributor = average_stars(repos)
    star_dict[contributor]=mean_contributor
    print("user {} has a mean of {}".format(contributor, mean_contributor))

    
# finally, sort the contributors

sorted_contributors = sorted(star_dict.items(), key=operator.itemgetter(1), reverse=True)
print(sorted_contributors)
```
